```python
import plotly.express as px
```

```python
columns_to_impute = [
    "GDP (current US$)",
    "GDP growth (annual %)",
    "GDP per capita (current US$)",
    "Inflation, consumer prices (annual %)",
    "Foreign direct investment, net inflows (% of GDP)",
    "Trade (% of GDP)",
    "Life expectancy at birth, total (years)",
    "Mortality rate, infant (per 1,000 live births)",
    "Access to electricity (% of population)",
    "People using at least basic drinking water services (% of population)",
    "People using at least basic sanitation services (% of population)",
    "Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita)",
    "PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)",
    "Renewable energy consumption (% of total final energy consumption)",
    "Forest area (% of land area)",
    "Control of Corruption: Percentile Rank",
    "Political Stability and Absence of Violence/Terrorism: Estimate",
    "Regulatory Quality: Estimate",
    "Rule of Law: Estimate",
    "Voice and Accountability: Estimate",
    "Individuals using the Internet (% of population)"
]

for col in columns_to_impute:
    df[col] = pd.to_numeric(df[col], errors='coerce')
for col in columns_to_impute:
    df[col] = df[col].fillna(df[col].mean())
```

```python
threshold = 0.7 * len(df)
columns_to_drop = [col for col in df.columns if df[col].isnull().sum() > threshold]
df = df.drop(columns=columns_to_drop, axis=1)
```

```python
df.info()
```

```python
num_data = df.select_dtypes(include=['number'])
corr = num_data.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True,cmap='crest')
plt.show()
```

```python
fig1 = px.scatter(df, x='GDP (current US$)', y='GDP per capita (current US$)',
                  title='GDP vs GDP per Capita', trendline='ols')
fig1.show()
```

```python
fig2 = px.scatter(df, x='GDP growth (annual %)', y='Unemployment, total (% of total labor force) (modeled ILO estimate)',
                  title='GDP Growth vs Unemployment', trendline='ols')
fig2.show()
```

```python
fig3 = px.scatter(df, x='Life expectancy at birth, total (years)', y='Mortality rate, infant (per 1,000 live births)',
                  title='Life Expectancy vs Infant Mortality', trendline='ols')
fig3.show()
```

```python
fig4 = px.scatter(df, x='Urban population (% of total population)', y='Access to electricity (% of population)',
                  title='Urban Population vs Access to Electricity', trendline='ols')
fig4.show()
```

```python
fig5 = px.line(df, x='Year', y='GDP growth (annual %)', color='Country',
              title='GDP Growth Over Time by Country')
fig5.show()
```

```python
fig6 = px.bar(df, x='Country', y='GDP per capita (current US$)',color='Year',
             title='GDP per Capita by Country')
fig6.show()
```

```python
fig7 = px.box(df, x='Country', y='GDP growth (annual %)',
             title='GDP Growth Distribution by Country')
fig7.show()
```

```python
fig8 = px.violin(df, x='Country', y='Life expectancy at birth, total (years)',
                title='Life Expectancy Distribution by Country')
fig8.show()
```

```python
fig9 = px.scatter_matrix(df, dimensions=['GDP (current US$)', 'GDP per capita (current US$)',
                                        'Life expectancy at birth, total (years)'],
                        color='Country', title='Pairwise Relationships')
fig9.show()
```

```python
fig10 = px.scatter(df, x='GDP (current US$)', y='GDP per capita (current US$)',
                 size='Population, total', color='Country',
                 title='Bubble Chart: GDP, GDP per Capita and Population')
fig10.show()
```

```python
fig11 = px.pie(df, values='GDP (current US$)', names='Country',
             title='GDP Contribution by Country')
fig11.show()
```
